# Dataset Understanding and Problem Framing

## Problem Statement

The objective of this project is to perform sentiment analysis on the IMDB movie reviews dataset. Given a movie review as text, the task is to classify it as expressing either positive or negative sentiment. The project further aims to compare multiple text vectorization techniques to analyze their impact on classification performance.

## Dataset Description

The IMDB dataset consists of 50,000 movie reviews labeled as either positive or negative. The dataset is balanced, making accuracy a suitable primary evaluation metric. Reviews vary in length and contain informal language, HTML tags, and punctuation, making it a suitable benchmark for natural language processing tasks.

# Text Preprocessing

## 'review' column preprocessing

```python
import re
import string
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
df['review']=df['review'].apply(clean_text)
```

```python
df.head()
```

```python
df['review'].iloc[0]
```

## 'sentiment' column preprocessing

```python
df['sentiment']=df['sentiment'].apply(lambda x: 1 if x=="positive" else 0 )
```

```python
df.head()
```

# Train–Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=df['review']
y=df['sentiment']
```

```python
X.shape,y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

# Text Vectorization

## Apply OHE (Binary BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ohe_vectorizer= CountVectorizer(binary=True,max_features=5000)
```

```python
X_train_ohe=ohe_vectorizer.fit_transform(X_train).toarray()
X_test_ohe=ohe_vectorizer.transform(X_test).toarray()
```

```python
X_train_ohe[0]
```

```python
X_test_ohe[0]
```

## Apply Bag of Words

```python
bow_vectorizer= CountVectorizer(max_features=5000,ngram_range=(1,2))
```

```python
X_train_bow=bow_vectorizer.fit_transform(X_train).toarray()
X_test_bow=bow_vectorizer.transform(X_test).toarray()
```

```python
X_train_bow[0]
```

```python
X_test_bow[0]
```

## Apply TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer= TfidfVectorizer(max_features=5000,ngram_range=(1,2))
```

```python
X_train_tfidf=tfidf_vectorizer.fit_transform(X_train).toarray()
X_test_tfidf=tfidf_vectorizer.transform(X_test).toarray()
```

```python
X_train_tfidf[0]
```

```python
X_test_tfidf[0]
```

# Model Training & Comparison

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression(max_iter=1000)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
model.fit(X_train_ohe,y_train)
y_pred=model.predict(X_test_ohe)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model.fit(X_train_bow,y_train)
y_pred=model.predict(X_test_bow)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model.fit(X_train_tfidf,y_train)
y_pred=model.predict(X_test_tfidf)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Word2Vec Embeddings

```python
token_train=[text.split() for text in X_train]
token_test=[text.split() for text in X_test]
```

```python
from gensim.models import Word2Vec
```

```python
w2v= Word2Vec(sentences=token_train,vector_size=100,window=5,min_count=2,workers=4)
```

```python
def sentence_vector(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)
```

```python
X_train_w2v = np.vstack([sentence_vector(t, w2v) for t in token_train])
X_test_w2v  = np.vstack([sentence_vector(t, w2v) for t in token_test])
```

```python
model.fit(X_train_w2v, y_train)
y_pred = model.predict(X_test_w2v)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Results and conclusions

## Final Results

The following table summarizes the performance of different text vectorizers applied to the IMDB movie reviews dataset using Logistic Regression as the classifier.

| Vectorization Method        | Accuracy |
|-----------------------------|----------|
| One-Hot Encoding (OHE)      | 0.8745   |
| Bag of Words (1–2 grams)    | 0.8732   |
| TF-IDF (unigrams)           | 0.8883   |
| **TF-IDF (1–2 grams)**      | **0.8920** |
| Word2Vec (Averaged Vectors) | 0.8483   |

Among all the methods, **TF-IDF with bigrams achieved the highest accuracy**, indicating its effectiveness in capturing both word importance and short-range contextual information relevant to sentiment analysis.

## Conclusion

In this project, multiple text representation techniques were evaluated to study their impact on sentiment classification performance on the IMDB movie reviews dataset. Simpler frequency-based methods such as One-Hot Encoding and Bag of Words provided reasonable performance but failed to effectively capture the importance of sentiment-bearing terms.

TF-IDF consistently outperformed these approaches by down-weighting commonly occurring but less informative words and emphasizing discriminative terms. The inclusion of bigrams further improved performance by capturing contextual patterns such as negation and intensifiers (e.g., *“not good”*, *“very bad”*), which are crucial for sentiment analysis tasks.

Although Word2Vec embeddings capture semantic relationships between words, the averaging approach used to obtain sentence vectors resulted in loss of word order and negation, leading to lower performance compared to TF-IDF-based methods. This highlights that for keyword-driven tasks like sentiment analysis, well-weighted frequency-based representations can outperform semantic embeddings.

Overall, this project demonstrates that the choice of text representation plays a critical role in NLP pipelines, and more complex models do not necessarily guarantee better performance without appropriate architectural support.
